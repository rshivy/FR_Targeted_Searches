```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib widget
```

```python
fig, ax = plt.subplots(3, 3, layout='constrained')

# X is uniform, log X is (?)

x = np.linspace(10 ** 5, 10 ** 10)
y = 1 / (10 ** 10 - 10 ** 5)
ax[0, 0].plot(x, [y] * len(x))
ax[0, 0].set_xlabel('$X$')
ax[0, 0].set_ylabel('PDF')

logx = np.linspace(5, 10)
y = (1 / (10 ** 10 - 10 ** 5)) * (10 ** logx) * np.log(10)

ax[0, 2].plot(logx, y)
ax[0, 2].set_xlabel('$\log_{10}X$')
ax[0, 2].set_ylabel('PDF')

# X is log-uniform, log X is uniform

x = np.linspace(10 ** 5, 10 ** 10)
y = (1 / (10 - 5)) * (1 / (x * np.log(10)))
ax[1, 0].plot(x, y)
ax[1, 0].set_xlabel('$X$')
ax[1, 0].set_ylabel('PDF')

logx = np.linspace(5, 10)
y = (1 / (10 - 5))

ax[1, 2].plot(logx, [y] * len(logx))
ax[1, 2].set_xlabel('$\log_{10}X$')
ax[1, 2].set_ylabel('PDF')

plt.show()
```
