```python
import matplotlib.pyplot as plt
import json
from astropy.coordinates import SkyCoord
%matplotlib inline
```

```python
full_priors = {}
with open('../Target_Priors/target_index.json', 'r') as f:
    index = json.load(f)
    for file_name in index.values():
        with open(f'../Target_Priors/{file_name}', 'r') as pfile:
            priors = json.load(pfile)
            full_priors[file_name[4:-5]] = priors
```

```python
full_priors.values()
```

```python
ras = [targ['RA'] for targ in full_priors.values()]
decs = [targ['DEC'] for targ in full_priors.values()]
coords = SkyCoord(ras, decs)
```

```python
ra_rad = coords.ra.radian
dec_rad = coords.dec.radian
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='mollweide')
ax.scatter(ra_rad, dec_rad, s=5)
ax.set_facecolor('w')
ax.grid()
```

```python
log10_dists = [targ['log10_dist'] for targ in full_priors.values()]
log10_freqs = [targ['log10_freq'] for targ in full_priors.values()]
```

```python
fig, ax = plt.subplots()
ax.hist(log10_dists)
ax.set_facecolor('w')
ax.grid()
```

```python
fig, ax = plt.subplots()
ax.hist(log10_freqs)
ax.grid()
```
